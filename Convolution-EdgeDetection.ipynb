```python
import matplotlib.image as mpim
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import convolve2d
```

```python
im=mpim.imread('lena.png')
```

```python
plt.imshow(im)
```

```python
# Sobel Filters for X direction
Hx=np.array([[-1.0,0.0,1.],[-2.0,0.0,2.],[-1.0,0.0,1.]])
```

```python
Hx.dtype
```

```python
# Sobel Filters for y direction
Hy=Hx.T
```

```python
bw=im.mean(axis=2)
```

```python
plt.imshow(bw,cmap='gray')
```

```python
Gx=convolve2d(bw,Hx)
```

```python
plt.imshow(Gx,cmap='gray')
```

```python
Gy=convolve2d(bw,Hy)

plt.imshow(Gy,cmap='gray')
```

```python
# Gx and Gy serve as vectors, so we can calculate magnitude and direction by
G=np.sqrt(Gx**2+Gy**2)
```

```python
plt.imshow(G,cmap='gray')
```

```python
# to plot the gradient's direction we find theta
theta=np.arctan2(Gy,Gx)
```

```python
plt.imshow(theta,cmap='gray')
```
