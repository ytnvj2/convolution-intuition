```python
import matplotlib.image as mpimage
import numpy as np
from scipy.signal import convolve2d
```

```python
import matplotlib.pyplot as plt
```

```python
im=mpimage.imread('lena.png')
```

```python
plt.imshow(im)
```

```python
bw=im.mean(axis=2)
```

```python
plt.imshow(bw,cmap='gray')
```

```python
W=np.zeros((20,20))
```

```python
for i in range(20):
    for j in range(20):
        d=(i-6.5)**2+(j-9.5)**2
        W[i,j]=np.exp(-d/200)
```

```python
plt.imshow(W,cmap='gray')
```

```python
blur=convolve2d(bw,W)
```

```python
plt.imshow(blur,cmap='gray')
```

```python
blur.shape
```

```python
bw.shape
```

```python
blur_same=convolve2d(bw,W,mode='same')
```

```python
plt.imshow(blur_same,cmap='gray')
```

```python
out3=np.zeros(im.shape)
```

```python
out3.shape
```

```python
for i in range(3):
    out3[:,:,i]=convolve2d(im[:,:,i],W,mode='same')
```

```python
plt.imshow(out3)
```

```python
plt.imshow(out3)
```
